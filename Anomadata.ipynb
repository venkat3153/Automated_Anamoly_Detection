**FEATURE ENGINEERING**

```python
# Handle invalid dates gracefully
df['time'] = pd.to_datetime(df['time'], format='%m-%d-%y %H:%M', errors='coerce')
```

```python
df.head()
```

```python
df.info()
```

```python
x=df.drop(columns=["y"],axis=1)
y=df["y"]
```

```python
y.value_counts()     
```

```python
# Extract components
x['year'] = x['time'].dt.year
x['month'] = x['time'].dt.month
x['day'] = x['time'].dt.day
x['hour'] = x['time'].dt.hour
x['minute'] = x['time'].dt.minute



```

```python
# Drop the original datetime column
x = x.drop(columns=['time'])
```

```python
x.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
xscaled=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
xscaled.head()
```

**MODEL TRAINING**

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Assume X is your feature matrix and y is your target variable
X_train, X_test, y_train, y_test = train_test_split(xscaled, y, test_size=0.2, random_state=42)

# Apply SMOTE to the training data
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

# Check new class distribution
from collections import Counter
print("Balanced class distribution:", Counter(y_train_balanced))
```

```python
model3=RandomForestClassifier()
model3.fit(X_train_balanced,y_train_balanced)
x_train_pred=model3.predict(X_train_balanced)
x_train_accuracy_score=accuracy_score(y_train_balanced,x_train_pred)
x_train_accuracy_score
```

**MODEL VALIDATION**

```python
x_test_pred=model3.predict(X_test)
x_test_accuracy_score=accuracy_score(y_test,x_test_pred)
x_test_accuracy_score
```

```python
import seaborn as sns
cm=confusion_matrix(y_test,x_test_pred)
hm=sns.heatmap(cm,annot=True,cbar=True)
hm
```
